```python
y = data["price"]
X = data[["bedrooms", "bathrooms", "area"]]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr_model = RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
param_gridrfr = {"max_depth" : [2, 5, 10, 15, 20, 25],
                 "max_features" : [3, "log2"],
                 "n_estimators" : [2, 4, 6, 8, 10, 12]}
```

```python
gridrfr = GridSearchCV(rfr_model, param_gridrfr, cv=3, scoring="neg_mean_absolute_error")
```

```python
gridrfr.fit(X_train, y_train)
```

```python
gridrfr.best_params_
```

```python
predictions = gridrfr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, predictions)
```
